# Linkedin

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
import random
import pandas as pd
import numpy as np
```

## Signin using linkedin credential (username and password) after running below cell

```python
url = 'https://www.linkedin.com/search/results/people/?keywords=software%20developer&origin=CLUSTER_EXPANSION&searchId=351f6dba-fbd9-4dfb-9308-c73248bd35e4&sid=stJ'
browser = webdriver.Chrome()
browser.get(url)
```

```python
links = []
```

```python
i = 2
while(len(links) < 100):
    k = 0
    profile_link = browser.find_elements(By.CSS_SELECTOR, "span")
    for link in profile_link:
        try:
            link = link.find_element(By.CSS_SELECTOR, "a.app-aware-link")
            link = link.get_attribute("href")
            if k%3 == 0:
                links.append(link)
            k+=1
        except:
            pass
    url = f"https://www.linkedin.com/search/results/people/?keywords=software%20developer&origin=CLUSTER_EXPANSION&page={i}&searchId=351f6dba-fbd9-4dfb-9308-c73248bd35e4&sid=JZH"
    browser.get(url)
    i += 1
```

```python
links
```

```python
names = []
current_positions = []
skill_links = []
profile_links = []
count = 1
for link in links:
    browser.get(link)
    profile_links.append(link)
    name = browser.find_element(By.CSS_SELECTOR, 'h1.text-heading-xlarge.inline.t-24.v-align-middle.break-words')
    names.append(name.text)
    current = browser.find_element(By.CSS_SELECTOR, 'div.text-body-medium.break-words')
    current_positions.append(current.text)
    print(f"<---{count}/100--->")
    print(name.text)
    count += 1
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    sleep(2)
    skill_lst = browser.find_elements(By.CSS_SELECTOR, 'a.optional-action-target-wrapper.artdeco-button.artdeco-button--tertiary.artdeco-button--standard.artdeco-button--2.artdeco-button--muted.inline-flex.justify-center.full-width.align-items-center.artdeco-button--fluid')
    lst = []
    for i in skill_lst:
        lst.append(i.get_attribute('href'))
    skill_links.append(lst)
```

```python
print(len(current_positions))
print(len(skill_links))
print(len(names))
```

```python
skills = []
count = 1
for skill_lst in skill_links:
    skill_link = ""
    for i in skill_lst:
        if 'skills' in i:
            skill_link = i
    if(skill_link == ""):
        skills.append('None')
        print(f"<---{count}/100--->")
        count += 1
        print("None")
        print('-----------------------------------------------------------------------------')
        continue
    browser.get(skill_link)
    try:
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    except:
        pass
    sleep(2)
    skill_str = ""
    skill = browser.find_elements(By.CSS_SELECTOR, "div.GGqtidempMjabazSDeNSVewmNsTMPkAZjuus.pvs-entity--padded.VxYeXnpWxxvbfarnTAGAGCKWSLQEQONNxxD")
    m = 0
    for i in skill:
        j = i.find_element(By.CSS_SELECTOR, 'span.visually-hidden')
        if j.text == '':
            if m == 1:
                break;
            m += 1
            continue
        skill_str = skill_str + j.text + ','
    print(f"<---{count}/100--->")
    count += 1
    print(skill_str[:-1])
    print('-----------------------------------------------------------------------------')
    skills.append(skill_str[:-1])
```

```python
df = pd.DataFrame(columns = ['Name', 'Current Position', 'Profile URL', 'Skills'])
df['Name'] = names
df['Current Position'] = current_positions
df['Profile URL'] = profile_links
df['Skills'] = skills
```

```python
df = df.iloc[:100]
df
```

```python
df.to_csv('linkedin_web_scrap.csv', index = False)
```

# Twitter

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
import random
import pandas as pd
import numpy as np
```

## Signin using twitter credential (username and password) after running below cell

```python
url = 'https://twitter.com/search?q=%23coding&src=typed_query'
browser = webdriver.Chrome()
browser.get(url)
```

```python
users_ids = []
tweets = []
likes = []
```

```python
while(len(likes) < 200):
    
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    rdm = random.randint(1, 5)
    sleep(rdm)
    
    user_id_and_like = browser.find_elements(By.CLASS_NAME,  "css-175oi2r.r-j5o65s.r-qklmqi.r-1adg3ll.r-1ny4l3l")
    tweet = browser.find_elements(By.CLASS_NAME,  "css-1rynq56.r-8akbws.r-krxsd3.r-dnmrzs.r-1udh08x.r-bcqeeo.r-qvutc0.r-37j5jr.r-a023e6.r-rjixqe.r-16dba41.r-bnwqim")
    
    ########################################################
    length = min(len(user_id_and_like), len(tweet))
    for i in range(length):
        try:
            my_str = user_id_and_like[i].text
            user_id = ""
            user_flag = 0
            for i in my_str:
                if i == '@':
                    user_flag = 1
                    continue
                if user_flag and i == '\n':
                    break;
                if user_flag:
                    user_id += i
            users_ids.append(user_id)
            #-------------------------------------------------#
            my_str = my_str[::-1]
            like = 0
            like_flag = 0
            for i in my_str:
                if i == '\n':
                    like_flag = 1
                    continue
                if like_flag == 1:
                    like = i
                    break
            likes.append(like)
        except:
            my_str = user_id_and_like[i].text
            user_id = ""
            user_flag = 0
            for i in my_str:
                if i == '@':
                    user_flag = 1
                    continue
                if user_flag and i == '\n':
                    break;
                if user_flag:
                    user_id += i
            users_ids.append(user_id)
            likes.append(0)
    ##########################################################
    
    for i in range(length):
        flag = 0
        try:
            my_str = tweet[i].text
            tweets.append(my_str)
            flag = 1
        except:
            tweets.append('')
            
```

```python
print(len(tweets))
print(len(users_ids))
print(len(likes))
```

```python
len(np.unique(np.array(users_ids)))
```

```python
k = 0
for tweet in tweets:
    tweets[k] = tweet.replace('\n', ' ')
    k += 1
    
k = 0
for like in likes:
    if like.isalpha() == 1:
        likes[k] = 0
    k += 1
```

```python
final_user_ids = []
final_tweets = []
final_likes = []
my_dict = {}
```

```python
k = 0
for user in users_ids:
    if user in my_dict:
        k+=1
        continue
    my_dict[user] = 0
    final_user_ids.append(user)
    final_tweets.append(tweets[k])
    final_likes.append(likes[k])
    k += 1
    if(len(final_likes) == 100):
        break
```

```python
df = pd.DataFrame(columns = ['Twitter handle', 'Likes', 'Tweet content'])
df['Twitter handle'] = final_user_ids
df['Tweet content'] = final_tweets
df['Likes'] = final_likes
```
